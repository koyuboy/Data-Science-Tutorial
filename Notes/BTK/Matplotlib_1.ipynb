```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
yasListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun yaşa göre değişim grafiği")


#plt.show()
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 **3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"y.-")
```

```python
plt.subplot(1,2,1) # 1 row olucak, 2 column olacak, şu an 1. grafiği çiziyorum
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) # 1 row olucak, 2 column olacak, şu an 2. grafiği çiziyorum
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

## Obje oluşturarak detaylı plot yapma

```python
benimFigure = plt.figure() #obje oluşturma
figureAxes = benimFigure.add_axes([0.2,0.2,0.7,0.5]) #(x kordinat,y kordinat,width,height)
#kordinatlar 2 grafik olduğunda işe yarar, aşşağıda 2 grafikli örneğe bakabilirsin
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni ismi")
figureAxes.set_ylabel("y ekseni ismi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])
eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("Büyük grafik başalığı")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("x ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("Küçük grafik başalığı")
```

## subplots ile figure objesi ve axes oluşturma 

```python
(benimFigure, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,"y-")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1,ncols=2)
for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")
plt.tight_layout()    # graifklerin arasında boşluk ayarlıyor
#benimFigure 
```

## Görsel Geliştirmeler legend, label etc.

```python
yeniFigur = plt.figure() # içi boşolabilir alttakileride yazabiliriz
#dpi =100 çözünürlük artıyo kaliteli oluyo arttıkça 
#figsize=(3,3)

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 **2,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 **3, label = "numpydizisi **3")
yeniEksen.legend(loc=6)
```

```python
yeniFigur.savefig("benimfigur.png",dpi=200) # grafiği bilgisayara kaydetme
```
