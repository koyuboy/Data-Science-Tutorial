# Çizgi Grafiği (Line Graph)

```python
ave = [3.2, 2.9, 3.5, 3.8]
years = [2018, 2019, 2020, 2021]
plt.plot(years,ave)
plt.ylabel("GNO")
plt.xlabel("Yıllar")
plt.show()
```

# Grafik Biçimlendirme

```python
ave = [3.2, 2.9, 3.5, 3.8]
ave2 = [2.7, 2.2, 1.8, 3.1]
years = [1, 2, 3, 4]
# plt.plot(y)-----plt.plot(x,y)
plt.plot(years,ave, label="std1")
plt.plot(years,ave2, "r^-.", label="std2")
# 3.parametre olarak stil verilir
# o->daire, *->yıldız, p->beşgen, s->kare, ^->üçgen veriler bu şekilde işaretlerinir çizgi yerine
# renk vermek için rengin baş harfi r->red,b->blue,y->yellow vb
# - >noktaları çizgiyle birleştirir, -. >tire ve nokta,-- >tire tire çizer, : > sadece noktalı çizgi

plt.legend(loc="lower left")
plt.ylabel("GNO")
plt.xlabel("Yıllar")
plt.axis([1, 4, 0, 4]) # xmin,xmax,ymin,ymax
plt.show()
```

```python
cases = np.arange(0, 5, 0.02)
```

```python
#2 şerli olarak tek bir plot() içinde verilebilir. Stil eklemek için her 3.ye stil yazabiliriz

plt.plot(cases,cases,"-r",cases,cases*2,"-.b",cases,cases**2,":y^")
plt.show()
#burda okadar çok nokta var ki noktalar kalın çizgi gibi duruyor ( sarıda)
#bunu çözmek için figure size 'ı büyütmeliyiz
```

# Pasta Grafiği (Pie Graph)

```python
shares = {"A":20,"B":10,"C":40,"D":30}
print(shares.keys())
print(shares.values())
#bunları kullanarak pasta grafiğine dönüştürücez
plt.pie(shares.values(), explode = [0,0,0.1,0],labels=shares.keys(),
       shadow = True, autopct="%1.1f%%" )
# ilk parametre grafiği oluşturen değerler
#explode grafikte elemanı öne çıkarmak için mesela C yi öne çıkakrdık
#labels=  A,B,C,D yi yazmak için
#shadow -> zorunlu değil gölgeli gözüksün diye
#autopct(auto percentage) -> labellara yüzde yazar  "%%" bu formatta içnie yazarsın
```

# Bar Graph'a Stil Ekleme

```python
people=["Monty","Ersin","Python","Jupyter"]
salaries=[2.5,1.1,3.9,1.8]

from matplotlib.ticker import FuncFormatter

def formatter(val, pos):
    return str(val) + " M $"

fig, ax = plt.subplots() 
ax.bar(people, salaries)
ax.yaxis.set_major_formatter(FuncFormatter(formatter))
ax.set_title("Maaş Grafiği")
```

# Scatter Plot

x ve y eksenindeki değerlerin korelesyonuna filan bakıyon, aralarında bağlantısına bakıyon

```python
ages=[3.5,2.5,1.5,5,4,1,2,3,4.5,2]
words=[45,30,10,100,80,3,35,55,67,15]

scatter = plt.scatter(ages, words,c= words,s=np.array(words)*10)
#c(colour), s(size)  aynı words ler aynı renkte oldu burda mesela
plt.colorbar(scatter) #en sağdaki bar için bir değişkene atanıp bu işlem yapılır
plt.show()
```

# Çoklu Plot

```python
ave = [3.2, 2.9, 3.5, 3.8]
ave2 = [2.7, 2.2, 1.8, 3.1]
years = [1, 2, 3, 4]

fig, axs = plt.subplots(2) # parametre kaç veri olucak sa burda 2 öğrenci için 2
axs[0].plot(years,ave)
axs[1].plot(years,ave2)
```

```python
fig ,axs = plt.subplots(2,2)
axs[0,0].plot(years,ave)
axs[0,1].plot(years,ave2)
axs[1,0].plot(years,np.array(ave)*(3/2))
#axs[1,1].plot(years,np.array(ave)*(2/3))
#böyle birini yazmasakda sorun yok boş gelir hata vermez
```
