# Mapping
## applymap()-->DataFrame
## map()-->Series
## apply()-->DataFrame+Series

apply() bir fonksiyonu bir dataframe'e veya seriye uygulamak için var
diğer ikisi lambda fonksyonu ile yapılır

```python
df[["NA_Sales","EU_Sales", "JP_Sales","Other_Sales"]].applymap(lambda line: str(line)+" $")
#birden çok kolon seçtiğimiz için dataframe oldu, tek kolon seçseydik applymap() hata verirdi çünkü o seris olur
```

```python
df["Name"].map(lambda line:line + " Game")
```

```python
def millonize(line):
    for j in range(len(line)):
        number = line[j]
        number = str(float(number)*1000000)[:-2]
        counter = 1
        result = ""
        for i in range(len(number)-1,-1,-1):
            if counter % 3 != 0:
                result = number[i]+result
            else:
                result ="," + number[i]+result  
            counter +=1
        if result[0] == ",":
            result = result[1:]
        line[j] = result
    return line
```

```python
df[["NA_Sales","EU_Sales", "JP_Sales","Other_Sales"]] = df[["NA_Sales","EU_Sales", "JP_Sales","Other_Sales"]].applymap(lambda line: str(line))
```

```python
df[["NA_Sales","EU_Sales", "JP_Sales","Other_Sales"]]=df[["NA_Sales","EU_Sales", "JP_Sales","Other_Sales"]].head(10).apply(millonize,axis = 1)
```

```python
df
```

```python
def dollarize(line):
    for col in ["NA_Sales","EU_Sales", "JP_Sales","Other_Sales"]:
        print(col + " = " + line[col] + " $")
```

```python
df[["NA_Sales","EU_Sales", "JP_Sales","Other_Sales"]].head(10).apply(dollarize,axis = 1)
```
